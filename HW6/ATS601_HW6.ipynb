```python
# plot initial wind fields for each component change of the zonal wind field
fig = plt.figure(figsize=(11,8),constrained_layout=True)
gs1 = gs.GridSpec(2,3,figure=fig)
plt.suptitle('Uwind Fields, $\zeta$ = 0',size=15)
### uwind plots
ax1 = fig.add_subplot(gs1[0,0])
plt.contourf(ds.lon,ds.lat,ds.ucomp[0,:,:],cmap=cmap,vmin=-30,vmax=30);
plt.title('Initial wind, No Change, + - +');labels()
ax2 = fig.add_subplot(gs1[0,1])
plt.contourf(ds.lon,ds.lat,ds1.ucomp[0,:,:],cmap=cmap,vmin=-30,vmax=30);
plt.title('First term sign reverse - - +');labels()
ax3 = fig.add_subplot(gs1[1,0])
plt.contourf(ds.lon,ds.lat,ds2.ucomp[0,:,:],cmap=cmap,vmin=-30,vmax=30);
plt.title('Second term sign reverse + + +');labels()
ax4 = fig.add_subplot(gs1[1,1])
plt.contourf(ds.lon,ds.lat,ds3.ucomp[0,:,:],cmap=cmap,vmin=-30,vmax=30);
plt.title('Third term sign reverse + - -');labels()
ax5 = fig.add_subplot(gs1[0, 2])
plt.contourf(ds.lon,ds.lat,ds4.ucomp[0,:,:],cmap=cmap,vmin=-30,vmax=30);
plt.title('All terms sign reversed - + -');labels()
ax6 = fig.add_subplot(gs1[1,2])
plt.contourf(ds.lon,ds.lat,ds5.ucomp[50,:,:],cmap=cmap,vmin=-30,vmax=30);
plt.title('All terms sign reversed - + -, w/ vorticity');labels()
# plt.savefig('/Users/tbarbero/Documents/FA21/601/HW6/figs/uwind_fields.png',dpi=300,bbox_inches='tight')
```
